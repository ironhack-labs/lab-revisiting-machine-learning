###### - Identify columns that over 85% missing values

```python
categorical["MAILCODE"].apply(lambda x:x.replace(" ","A"))
categorical=categorical.replace(" ",np.nan)
```

```python
nulls_df=pd.DataFrame(categorical.isna().sum()/len(categorical))
nulls_df.columns=["nulls_percentage"]
nulls_df
```

```python
nulls_df[nulls_df["nulls_percentage"]>0.85].sort_values(by=["nulls_percentage"],ascending=False)
```

```python
cols_to_drop=list(nulls_df[nulls_df["nulls_percentage"]>0.25].sort_values(by=["nulls_percentage"],ascending=False).index)
cols_to_drop
```

###### - Remove those columns from the dataframe

```python
categorical=categorical.drop(columns=cols_to_drop)
categorical
```

###### - Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, and "other" for all the rest

```python
categorical["GENDER"].value_counts()
```

```python
variables=["U","J","C","A"]

for var in variables:
    categorical['GENDER'] = categorical['GENDER'].replace(var,"other")
    
categorical["GENDER"].value_counts()
```
