```python
import pandas as pd
```

```python
data = pd.read_csv('categorical_1.csv')
```

```python
data
```

```python
#Check for null values in all the columns

pd.set_option('display.max_rows',None)
nulls_df = pd.DataFrame(data.isna().sum()/len(data))
```

```python
nulls_df
```

```python
#Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. 
#We will append this list and then drop all the columns in this list later:

drop_list = []

drop_list.append('OSOURCE')
drop_list.append('ZIP')
```

```python
drop_list
```

```python
nulls_df.columns=['nulls_percentage']
```

```python
#Identify columns that over 85% missing values
```

```python
cols_to_drop = nulls_df[nulls_df['nulls_percentage']>.85].sort_values(
    by=['nulls_percentage'],ascending=False).index
```

```python
cols_to_drop
```

```python
#Remove those columns from the dataframe
```

```python
cols_to_drop = [col for col in cols_to_drop if col not in drop_list]

data = data.drop(columns=cols_to_drop)
print(data.shape)
```

```python
data = data.drop(columns=drop_list)
print(data.shape)
```

```python
data.head()
```

```python
#Reduce the number of categories in the column GENDER. The column should only have either "M" for males, "F" for females, 
#and "other" for all the rest
```

```python
data['GENDER']
```

```python
print(data['GENDER'].value_counts())
data['GENDER'] = data['GENDER'].fillna('F')
```

```python
values_to_replace = {'U': 'other', 'J': 'other', 'C': 'other', 'A': 'other'}

# Replacing the values in the 'GENDER' column
data['GENDER'] = data['GENDER'].replace(values_to_replace)
```
