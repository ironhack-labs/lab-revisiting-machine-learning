```python
# Libaries

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\ljant\Desktop\Ironhack\15_Intro_ML\learningSet.txt")
data
```

```python
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
# Clean categorcial columns
#1. Check for null values in all the columns
sum_na = pd.DataFrame(data.isna().sum()/len(data)).reset_index()
sum_na.columns = ['column_name', 'nans_percentage']
sum_na.sort_values(by = ['nans_percentage'], ascending = False)
```

```python
'''
2. Exclude the following variables by looking at the definitions. Create a new empty list called drop_list. 
We will append this list and then drop all the columns in this list later:

OSOURCE - symbol definitions not provided, too many categories
ZIP CODE - we are including state already
'''

drop_list = ['OSOURCE', 'ZIP'] 
drop_list
```

```python
# 3. Identify columns that over 25% missing values
threshold = 0.25
condition = sum_na['nans_percentage'] > threshold
columns_above_threshold = sum_na[condition]
columns_above_threshold

drop_list.extend(columns_above_threshold['column_name'].tolist())
drop_list
```

```python
# 4. Remove those columns from the dataframe
data.drop(drop_list, axis=1, inplace=True)
data
```

```python
'''
5. Reduce the number of categories in the column GENDER. The column should only have either "M" for males, 
"F" for females, and "other" for all the rest
'''

# Removing NaNs from this column first
print(data['GENDER'].value_counts())
data['GENDER'] = data['GENDER'].replace({' ': np.nan})
# 
data['GENDER'] = data['GENDER'].fillna('other')
print(data['GENDER'].value_counts())

# adding all the other categories to 'other'
data['GENDER'] = data['GENDER'].replace({'C': 'other', 'U': 'other', 'J': 'other', 'A': 'other'})
print(data['GENDER'].value_counts())
```

```python
# Safe new data file
data.to_csv('learningSet_2.txt', index=False)
```
